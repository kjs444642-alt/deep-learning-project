# 프로젝트 제안서

## 1. 프로젝트 개요

* 프로젝트 이름: 콘크리트 균열 자동 탐지 시스템
* 작성자: 김준서
* 작성일: 4/5

## 2. 배경 및 필요성

* 건물의 균열은 구조적 안전과 직결되는 중요한 문제로, 조기 발견이 필요하다. 그러나 기존의 점검 방식은 사람이 직접 확인해야 하므로 시간과 비용이 많이 소요되며, 주관적인 판단이 개입될 수 있다.
따라서 딥러닝 기반 이미지 분석을 활용하여 균열을 자동으로 탐지함으로써 보다 효율적이고 객관적인 점검이 가능하다.

## 3. 데이터셋

* 본 프로젝트에서는 Kaggle에서 제공하는 "Concrete Crack Images for Classification" 데이터셋을 활용한다.  

* 해당 데이터셋은 콘크리트 표면 이미지를 기반으로 구성되어 있으며, 균열이 있는 이미지와 균열이 없는 이미지로 구분되어 있다.  

- Positive (균열 있음): 약 20,000장  
- Negative (균열 없음): 약 20,000장  

* 총 약 40,000장의 이미지로 구성되어 있으며, 클래스 간 데이터 수가 균형을 이루고 있어 딥러닝 모델 학습에 적합하다. [링크 텍스트](https://www.kaggle.com/datasets/arnavr10880/concrete-crack-images-for-classification)

## 4.목표

* 본 프로젝트의 목표는 콘크리트 표면 이미지를 입력받아 균열 여부를 자동으로 판별하는 딥러닝 모델을 구현하는 것이다.  
CNN(합성곱 신경망)을 활용하여 이미지의 특징을 학습하고, 이를 기반으로 균열(Positive)과 비균열(Negative)을 정확하게 분류하는 것을 목표로 한다.  
또한 학습된 모델의 성능을 평가하여 실제 건축물 점검에 활용 가능한 수준의 정확도를 확보하는 것을 목표로 한다.

## 5. 주요 기능 및 특징

본 프로젝트의 주요 기능 및 특징은 다음과 같다.  

1) 균열 이미지 자동 판별 기능  
콘크리트 표면 이미지를 입력받아 균열 여부를 자동으로 판별한다.  

2) 이미지 분류 기능  
입력된 이미지를 균열(Positive)과 비균열(Negative)로 분류한다.  

3) 빠른 점검 가능  
대량의 이미지를 빠르게 처리하여 효율적인 점검이 가능하다.  

4) 일관된 판단 제공  
사람의 주관이 개입되지 않아 일정한 기준으로 결과를 제공한다.  

5) 유지보수 활용 가능  
건축물의 균열 여부를 사전에 파악하여 유지보수에 활용할 수 있다.

## 6. 기대 효과

* 딥러닝 기반 균열 탐지 시스템을 통해 건축물 점검의 효율성과 정확도를 향상시킬 수 있다. 또한 사람의 주관을 배제한 객관적인 판단이 가능하며, 유지보수 비용 절감 및 안전성 확보에 기여할 수 있다.

## 10. 기존 프로젝트의 문제점 개선

기존의 균열 탐지 모델은 주어진 이미지 데이터만을 그대로 학습하여, 다양한 환경(회전, 밝기 변화, 확대/축소, 좌우 반전 등)에 대한 대응이 부족한 한계가 있다.  

이를 개선하기 위해 데이터 증강(Data Augmentation)을 적용하여 이미지의 형태를 다양하게 변형하고, 다양한 상황에서도 균열을 정확히 인식할 수 있도록 개선한다.


## 11. 실제 환경 반영 방안

실제 환경에서는 촬영 각도, 조명, 거리 등이 일정하지 않기 때문에 다양한 형태의 이미지가 입력될 수 있다.  

이를 반영하기 위해 회전, 밝기 변화, 확대 및 좌우 반전 등의 데이터 증강 기법을 적용하여, 다양한 조건에서도 안정적으로 작동하는 모델을 구현한다.


## 12. 성능 개선 방안

데이터 증강을 통해 학습 데이터의 다양성을 확보함으로써 모델의 일반화 성능을 향상시킨다.

## 13. CNN 아키텍쳐

본 모델의 CNN 아키텍처는 Conv2D, Batch Normalization, ReLU, MaxPooling Layer를 반복적으로 구성하여 이미지의 특징을 추출하도록 설계하였다.  

이후 Adaptive Average Pooling과 Dense Layer를 통해 추출된 특징을 기반으로 균열 여부를 최종 분류한다.
