# 참고 프로젝트 분석

## 1. 참고 프로젝트 개요
본 프로젝트는 건물 외벽의 이미지를 활용하여 균열 여부를 자동으로 판별하는 딥러닝 기반 이미지 분류 모델을 구현하는 것을 목표로 한다.  

이미지 데이터는 Kaggle의 "Concrete Crack Images for Classification" 데이터셋을 활용하며, 균열이 있는 이미지(Positive)와 균열이 없는 이미지(Negative)로 구성되어 있다.  

CNN(합성곱 신경망)을 이용하여 이미지의 특징을 학습하고, 이를 기반으로 균열 여부를 분류하는 모델을 구축한다.

---

## 2. 선택 이유
첫째, 건물 균열 탐지는 실제 건축 및 안전 분야에서 중요한 문제이기 때문이다.  
둘째, 이미지 기반 데이터로 딥러닝(CNN)을 적용하기에 적합한 주제이기 때문이다.  
셋째, 데이터셋이 충분히 크고(약 40,000장) 라벨이 명확하여 모델 학습에 유리하기 때문이다.  

또한, 결과가 시각적으로 확인 가능하여 프로젝트 결과를 직관적으로 이해할 수 있다는 장점이 있다.

---

## 3. 해당 프로젝트 분석 내용

### (1) 데이터 분석
- 데이터 출처: Kaggle  
- 데이터셋: Concrete Crack Images for Classification  
- 총 약 40,000장의 이미지  
- Positive (균열 있음): 약 20,000장  
- Negative (균열 없음): 약 20,000장  

데이터가 균형 잡혀 있어(class imbalance 없음) 모델 학습 시 편향이 적다.

---

### (2) 모델 분석
본 프로젝트에서는 CNN(Convolutional Neural Network)을 활용한다.  

CNN은 이미지의 특징(패턴, 경계, 질감 등)을 자동으로 추출하는 데 강점이 있어 균열 탐지 문제에 적합하다.  

기본적인 Conv2D(선, 모양, 패턴(균열 같은 것) 찾는 역할), MaxPooling(이미지를 작게 줄이면서 중요한 정보만 남김), Dense Layer(균열인지 아닌지 결정하는 부분)로 구성된 모델을 사용하여 이미지 분류를 수행한다.

---

### (3) 기대 결과
- 균열 여부를 자동으로 분류하는 모델 구현  
- 정확도 약 90% 이상의 성능 기대  
- 건물 안전 점검 자동화 가능성 확인  

---

### (4) 한계점
- 실제 환경에서는 다양한 노이즈(조명, 오염 등)가 존재할 수 있음  
- 단순 분류 모델이므로 균열의 위치까지는 정확히 파악하기 어려움
